```python
# Переменные
var("p1,p2,q1,q2")
# Постоянные
var("a,b,f,g")
H=(p1^2+p2^2)/2 + (q1^2+q2^2)^2 + a*q1^2 + b*q2^2
F=(q1*p2-q2*p1)^2+(b-a)*(p1^2+2*q1^4+2*q1^2*q2^2+2*a*q1^2)
G=(q1*p2-q2*p1)^2+(a-b)*(p2^2+2*q2^4+2*q1^2*q2^2+2*b*q2^2)
```

Шаг 1. Ищем
$$
u= \frac{\partial p_1}{\partial q_2}
$$
Для этого ищем уравнение
$$
F_1(q_1, q_2, p_1)=0
$$

```python
k=PolynomialRing(QQ,[p2,p1,q2,q1,a,b,f,g], order='lex')
J=k*[F==f,G==g]
B=J.groebner_basis()
```

```python
Eqp1=QQ[q1,q2,a,b,f,g][p1](B[-1])
Eqp1
```

```python
len(QQ[q1,q2,a,b,f,g,p1](Eqp1).monomials())
```

Теперь берем
$$
F(q_1,q_2,p_1)=0
$$
и дифференцируем по $q_2$
$$
F_{q_2}+ F_{p_1} u=0
$$
Отсюда
$$
u=-F_{q_2}/F_{p_1}
$$

```python
u=-diff(SR(Eqp1),q2)/diff(SR(Eqp1),p1)
u
```

Шаг 2. Ищем 
$$
v= \frac{\partial p_2}{\partial q_1}
$$
В определении кольца переставляем $p_1$ и $p_2$ и получаем
$$
F_2(q_1,q_2,p_2)=0
$$

```python
k=PolynomialRing(QQ,[p1,p2,q2,q1,a,b,f,g], order='lex')
J=k*[F==f,G==g]
B=J.groebner_basis()
```

```python
Eqp2=QQ[q1,q2,a,b,f,g][p2](B[-1])
Eqp2
```

```python
v=-diff(SR(Eqp2),q1)/diff(SR(Eqp2),p2)
v
```

```python
w=SR(u-v).numerator()
w
```
