```python
# Переменные
var("p1,p2,q1,q2,c2")
# Постоянные
var("a,b,f,g,h")
H=(p1^2+p2^2)/2 + (q1^2+q2^2)^2 + a*q1^2 + b*q2^2
F=(q1*p2-q2*p1)^2+(b-a)*(p1^2+2*q1^4+2*q1^2*q2^2+2*a*q1^2)
G=(q1*p2-q2*p1)^2+(a-b)*(p2^2+2*q2^4+2*q1^2*q2^2+2*b*q2^2)
```

Система Ванеке имеет вид:

$$\dot{q_1}=\frac{\partial H}{\partial p_1}=p_1$$

$$\dot{q_2}=\frac{\partial H}{\partial p_2}=p_2$$

$$\dot{p_1}=-\frac{\partial H}{\partial q_1}$$

$$\dot{p_2}=-\frac{\partial H}{\partial q_2}$$

Рассмотрим случай $q_2=Const$. Тогда $p_2=0$. Тогда из последнего уравнения получаем $q_1=Const$:

```python
solve(diff(H,q2)==0,q1)
```

# Вопрос


Можем ли мы заменить систему Ванеке на систему из двух уравнений, в которых правые части получены как следствие из $F=Const,G=Const$? (Нулевой шаг теоремы Лиувилля по самому Лиувиллю и по Уиттекеру -- найти формулы для импульсов, у нас они есть.) Получится система

$$\dot{q_1}=p_1(q_1,q_2,f,g)$$

$$\dot{q_2}=p_2(q_1,q_2,f,g)$$

### Можем ли мы добавить к ней $q2=Const,p2=0$?

Так получается уравнение Ахмеддины:

$$\dot{q_1}=p_1(q_1;q_2=Const,f,g,a,b)$$

Данное уравнение интегрируется в квадратурах Ахмеддины:

$$\int\limits_{0}^{q_1(t)}\frac{dq_1}{p_1(q_1;params)}=t$$


Для интегрирования данного уравнения не будем выписывать полученную нами формулу для $p1(q1,q2,f,g)$. Учтём, что выполнены уравнения поверхности совместного уровня интегралов $F=f,G=g$. Выполним подстановку Ахмеддины $q2=Const,p2=0$.

# Теорема Лиувилля, точная форма, 2025

Шаг 1. Ищем
$$
u= \frac{\partial p_1}{\partial q_2}
$$
Для этого ищем уравнение
$$
F_1(q_1, q_2, p_1)=0
$$

```python
k=PolynomialRing(QQ,[p2,p1,q2,q1,a,b,f,g], order='lex')
J=k*[F==f,G==g]
B=J.groebner_basis()
```

```python
Eqp1=QQ[q1,q2,a,b,f,g][p1](B[-1])
Eqp1
```

Теперь берем
$$
F(q_1,q_2,p_1)=0
$$
и дифференцируем по $q_2$
$$
F_{q_2}+ F_{p_1} u=0
$$
Отсюда
$$
u=-F_{q_2}/F_{p_1}
$$

```python
u=-diff(SR(Eqp1),q2)/diff(SR(Eqp1),p1)
u
```

Шаг 2. Ищем 
$$
v= \frac{\partial p_2}{\partial q_1}
$$
В определении кольца переставляем $p_1$ и $p_2$ и получаем
$$
F_2(q_1,q_2,p_2)=0
$$

```python
k=PolynomialRing(QQ,[p1,p2,q2,q1,a,b,f,g], order='lex')
J=k*[F==f,G==g]
B=J.groebner_basis()
```

```python
Eqp2=QQ[q1,q2,a,b,f,g][p2](B[-1])
Eqp2
```

```python
len(QQ[q1,q2,a,b,f,g,p2](Eqp2).monomials())
```

```python
v=-diff(SR(Eqp2),q1)/diff(SR(Eqp2),p2)
v
```

Шаг 3. Нам нужно проверить
$$
\frac{\partial p_1}{\partial q_2}=\frac{\partial p_2}{\partial q_1},
$$
то есть верно ли равенство
$$
u(q_1,q_2,p_1)=v(q_1,q_2,p_2)
$$
при всех $p_1,p_2,q_1,q_2$. Верно ли, что числитель разность $u-v$ принадлежит иделалу $J$, порожденному элементами $F-f, G-g$?

```python
bool(u==v)
```

```python
w=SR(u-v).numerator()
w
```

```python
len(QQ[a, b, f, g, p1, p2, q1, q2](w).monomials())
```

```python
k=PolynomialRing(QQ,[p1,p2,q2,q1,a,b,f,g])
J=k*[F==f,G==g]
```
